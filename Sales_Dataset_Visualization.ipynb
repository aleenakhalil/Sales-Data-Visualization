##Visualizing Sales Data (Kaggle Sales Dataset)

Step 1: Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Loading the Dataset

```python
sales_df = pd.read_csv('sales_data.csv', encoding='ISO-8859-1')
```

Identifying Missing Values

```python
sales_df.isnull().sum()
```

Converting ORDERDATE to Datetime

```python
sales_df['ORDERDATE'] = pd.to_datetime(sales_df['ORDERDATE'])
```

Step 4: Handling Missing Data

Instead of replacing missing values in columns like TERRITORY, STATE, and POSTALCODE with methods such as mean or median (which could introduce inaccuracies), placeholder values such as 'Unknown' or '00000' have been assigned.

Since ADDRESSLINE2 is usually optional, it can be safely dropped or left as is without significantly affecting the dataset.

```python
sales_df['STATE'] = sales_df['STATE'].fillna('Unknown')
sales_df['POSTALCODE'] = sales_df['POSTALCODE'].fillna('00000')
sales_df['TERRITORY'] = sales_df['TERRITORY'].fillna('Unknown')
```

Step 4: Visualizations

```python
sales_df['Month'] = sales_df['ORDERDATE'].dt.to_period('M')
monthly_sales = sales_df.groupby('Month')['SALES'].sum()

plt.figure(figsize=(8,6))
monthly_sales.plot(kind='line', color='orange', linewidth=2)
plt.title('Total Sales Over Time (Monthly)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
country_sales = sales_df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=country_sales.index, y=country_sales.values, palette='husl')
plt.title('Sales by Country', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
product_line_sales = sales_df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=product_line_sales.index, y=product_line_sales.values, palette='viridis')
plt.title('Sales by Product Line', fontsize=16)
plt.xlabel('Product Line', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
deal_size_sales = sales_df.groupby('DEALSIZE')['SALES'].sum()

plt.figure(figsize=(8,6))
sns.barplot(x=deal_size_sales.index, y=deal_size_sales.values, palette='Blues')
plt.title('Sales by Deal Size', fontsize=16)
plt.xlabel('Deal Size', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
corr = sales_df[['SALES', 'QUANTITYORDERED', 'PRICEEACH', 'MSRP']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='BuGn', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='QUANTITYORDERED', y='SALES', data=sales_df, color='green', alpha=0.6)
plt.title('Sales vs. Quantity Ordered', fontsize=16)
plt.xlabel('Quantity Ordered', fontsize=14)
plt.ylabel('Sales ($)', fontsize=14)
plt.tight_layout()
plt.show()
```
